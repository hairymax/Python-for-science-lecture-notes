Наиболее общей процедурой интегрирования является [`scipy.integrate.quad()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html#scipy.integrate.quad). Вычислим: $\int\limits_0^\frac{\pi}{2} sin(x)dx$  

````python
from scipy.integrate import quad
res, err = quad(np.sin, 0, np.pi/2)
np.allclose(res, 1)   # res - результат, который для этого интеграла должен быть близок к 1
````

````python
np.allclose(err, 1 - res)  # err - ошибка расчётов
````

### Интегрирование дифференциальных уравнений

`scipy.integrate` так же содержит процедуры для интегрирования ОДУ. На практике  [`scipy.integrate.odeint()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.odeint.html#scipy.integrate.odeint) решает ОДУ следующего вида:
```
dy/dt = rhs(y1, y2, .., t0,...)
``` 
Для начала давайте решим следующее ОДУ: 
$$\frac{dy}{dt} = -2y$$ 
для $t$ в диапазоне значений $t = 0..4$ с начальным условием $y(t=0)=1$.

Сперва необходимо определить функцию, которая вычисляет производную.

````python
def calc_derivative(ypos, time):
    return -2 * ypos
````

Затем вычислить `y` как функцию от времени.

````python
from scipy.integrate import odeint
time_vec = np.linspace(0, 4, 40)
y = odeint(calc_derivative, y0=1, t=time_vec)

plt.xlabel("t: время")
plt.ylabel("y: положение")
plt.plot(time_vec, y)
plt.show()
````

Проинтегрируем более сложное ОДУ: уравнение затухающего пружинного маятника. 
Положение прикреплённой к пружине массы описывается уравнением 2-го порядка 
$$y''+ 2\epsilon\omega_{0} y' + \omega^2_{0} y = 0 ;$$ 
где 
$$\omega^2_{0} = \frac{k}{m}, \;\;\;\; \epsilon = \frac{c}{2m\omega_{0}}, $$ 
$k$ - жёсткость пружины, $c$ - коэффициент затухания. 

Установим значения

````python
mass = 0.5  # kg
kspring = 4  # N/m
cviscous = 0.4  # N s/m
````

````python
eps = cviscous / (2 * mass * np.sqrt(kspring/mass))
omega = np.sqrt(kspring / mass)
````

Колебания затухающие, поскольку: 

````python
eps < 1
````

Для `odient()` уравнение 2-го порядка необходимо преобразовать в систему из двух уравнений первого порядка для вектора $Y=(y,y')$.  
Функция вычисления скорости и ускорения.


````python
def calc_deri(yvec, time, eps, omega):
    return (yvec[1], -2.0 * eps * omega * yvec[1] - omega **2 * yvec[0]) 
````

````python
time_vec = np.linspace(0, 10, 100)
yinit = (1, 0)
yarr = odeint(calc_deri, yinit, time_vec, args=(eps, omega))

plt.plot(time_vec, yarr)
plt.legend(["$y$","$y'$"]);
````

`scipy.integrate.odeint()` использует LSODA (Livermore Solver for Ordinary Differential equations with Automatic method switching for stiff and non-stiff problems) / (Решатель Ливермора для обыкновенных дифференциальных уравнений с автоматическим переключением методов для жестких и нежестких задач). Более подробная информация см. [ODEPACK Fortran library](https://people.sc.fsu.edu/~jburkardt/f77_src/odepack/odepack.html ). 
 

**См. также** Дифференциальные уравнения в частных производных (Partial Differential Equations)
В *Scipy* нет решателя дифференциальных уравнений в частных производных (PDE). Но доступны некоторые пакеты Python для решения PDE, такие как [`fipy`](http://www.ctcms.nist.gov/fipy/) или [`SfePy`](http://sfepy.org/doc/).

## Быстрое преобразование Фурье: [`scipy.fftpack`](https://docs.scipy.org/doc/scipy/reference/fftpack.html#module-scipy.fftpack)



Модуль `scipy.fftpack` позволяет производить быстрые преобразования Фурье (БПФ) / (Fast Fourier transform - FFT) и содержит утилиты для их обработки. 

Главными функциями являются: 
* [`scipy.fftpack.fft()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.fftpack.fft.html#scipy.fftpack.fft) для вычисления FFT;
* [`scipy.fftpack.fftfreq()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.fftpack.fftfreq.html#scipy.fftpack.fftfreq) для генерации частот дискретизации;
* [`scipy.fftpack.ifft()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.fftpack.ifft.html#scipy.fftpack.ifft) для вычисления обратного Фурье-преобразования.


В качестве примера приведём (зашумленный) входной сигнал (`sig`) и его БПФ:

````python
from scipy import fftpack

# устанавливаем состояние генератора случайных чисел
np.random.seed(1234)

time_step = 0.02
period = 5.

time_vec = np.arange(0, 20, time_step)
sig = (np.sin(2 * np.pi / period * time_vec)
       + 0.5 * np.random.randn(time_vec.size))

plt.figure(figsize=(12, 5))
plt.plot(time_vec, sig, label='Оригинальный сигнал')
plt.show()
````

````python
# БПФ сигнала
sig_fft = fftpack.fft(sig)
# И мощность (sig_fft имеет тип данных complex dtype)
power = np.abs(sig_fft)**2
# Соответствующие частоты
sample_freq = fftpack.fftfreq(sig.size, d=time_step)
# График мощности БПФ
plt.figure(figsize=(10, 5))
plt.plot(sample_freq, power)
plt.xlabel('Частота, Гц')
plt.ylabel('Мощность')
# Поиск пиковой частоты: учитываем только на положительные частоты
pos_mask = np.where(sample_freq > 0)
freqs = sample_freq[pos_mask]
peak_freq = freqs[power[pos_mask].argmax()]
# Проверим, что она действительно соответствует частоте, 
# с которой мы генерируем сигнал
print(np.allclose(peak_freq, 1./period))
# график для отображения пиковой частоты
axes = plt.axes([0.55, 0.3, 0.3, 0.5])
plt.title('Пиковая частота')
plt.plot(freqs[:8], power[:8])
plt.setp(axes, yticks=[]);
# scipy.signal.find_peaks_cwt также может быть использован для более продвинутого обнаружение пиков
````

Поскольку сигнал вещественный, то преобразование Фурье симметрично.

Пик сигнала возможно найти с помощью `freqs[power.argmax()]`

Замена компонент Фурье-преобразования выше этой частоты на ноль и обратное преобразование FFT с помощью `scipy.fftpack.ifft()` даст нам отфильтрованный сигнал. 

````python
# фильтрация
high_freq_fft = sig_fft.copy()
high_freq_fft[np.abs(sample_freq) > peak_freq] = 0
filtered_sig = fftpack.ifft(high_freq_fft)
# график
plt.figure(figsize=(12, 5))
plt.plot(time_vec, sig, label='Оригинальный сигнал')
plt.plot(time_vec, filtered_sig.real, linewidth=3, label='Отфильтрованный сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда')
plt.legend(loc='best')
plt.show()
````

**`numpy.fft`**  
Библиотека `Numpy` также имеет реализацию FFT (`numpy.fft`). Однако реализация `scipy` предпочтительнее, поскольку в ней используются более эффективные базовые реализации.

## Обработка сигналов: [`scipy.signal`](https://docs.scipy.org/doc/scipy/reference/signal.html#module-scipy.signal)

`scipy.signal` предназначен для типичной обработки сигналов: одномерных, сигналов с регулярной дискретизацией.

**Ресэмплирование** [`scipy.signal.resample()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.resample.html#scipy.signal.resample): повторное сэмплирование сигнала в n точек с использованием БПФ (FFT).

````python
from scipy import signal
# исходный сигнал 20 точек
t = np.linspace(0, 10, 20, endpoint=False)
x = np.sin(-t**2/5.0)
# ресемплированный 100 точек
x_resampled = signal.resample(x, 100)
tnew = np.linspace(0, 10, 100, endpoint=False)
# график
plt.figure(figsize=(10, 5))
plt.plot(t, x, 'o-', label = 'Исходный сигнал') 
plt.plot(tnew, x_resampled, '.-', label = 'Ресэмплированный сигнал')
plt.legend();
````

Полученный ресемлированием набор точек отличается от того, что мы получили бы при интерполяции с помощью `scipy.interpolate`, поскольку она применяется только к регулярно распределённым данным.

**Удаление тренда** [`scipy.signal.detrend()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.detrend.html#scipy.signal.detrend): удаление линейного тренда из сигнала:

````python
from scipy import signal
t = np.linspace(0, 5, 100)
x = t + np.random.normal(size=100)
x_detrended = signal.detrend(x)
# график
plt.figure(figsize=(10, 5))
plt.plot(t, x, label = 'x')
plt.plot(t, x_detrended, label = 'x без тренда')
plt.legend();
````

**Фильтрация**: Для нелинейной фильтрации `scipy.signal` имеет фильтрацию (медианный фильтр `scipy.signal.medfilt()`, Винера `scipy.signal.wiener()`).

`scipy.signal` также имеет полноценный набор инструментов для проектирования линейных фильтров (фильтры с конечным и бесконечным откликом), но это выходит за рамки данного руководства.

**Спектральный анализ**: [`scipy.signal.spectrogram()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.spectrogram.html#scipy.signal.spectrogram) вычисляет спектрограмму – частотные спектры в последовательных временных окнах –, в то время как [`scipy.signal.welch()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.welch.html#scipy.signal.welch) вычисляет плотность спектра мощности (PSD).

### Геометрические преобразования изображений

Повороты, изменение разрешения, ..

````python
from scipy import misc
from scipy import ndimage

face = misc.face(gray=True)
shifted_face = ndimage.shift(face, (50, 50))
shifted_face2 = ndimage.shift(face, (50, 50), mode='nearest')
rotated_face = ndimage.rotate(face, 30)
cropped_face = face[50:-50, 50:-50]
zoomed_face = ndimage.zoom(face, 2)
zoomed_face.shape
````

````python
imgs = [shifted_face, shifted_face2, rotated_face, cropped_face, zoomed_face]

fig = plt.figure(figsize=(15,5))
for i in range(len(imgs)):
    fig.add_subplot(1,5,i+1)
    plt.imshow(imgs[i], cmap=plt.cm.gray)
    plt.axis('off')
````

### Фильтрация изображений

Создайте шумное лицо:

````python
from scipy import misc
face = misc.face(gray=True)
face = face[:512, -512:]
noisy_face = np.copy(face).astype(float)
noisy_face += face.std() * 0.5 * np.random.standard_normal(face.shape)
````

Применяйте к нему различные фильтры:

````python
from scipy import signal
blurred_face = ndimage.gaussian_filter(noisy_face, sigma=3)
median_face = ndimage.median_filter(noisy_face, size=5)
wiener_face = signal.wiener(noisy_face, (5, 5))
````

````python
imgs = [noisy_face, blurred_face, median_face, wiener_face]
names = ['noisy', 'blurred', 'median', 'wiener']

fig = plt.figure(figsize=(15,5))
for i in range(len(imgs)):
    fig.add_subplot(1,4,i+1)
    plt.imshow(imgs[i], cmap=plt.cm.gray)
    plt.title(names[i])
    plt.axis('off')
````

К изображениям могут быть применены другие фильтры из `scipy.ndimage.filters` и `scipy.signal`.

### **Упражнение**

Проверьте, что операция открытия эквивалентна последовательному выполнению операций эрозии и расширения.

**Закрытие**: [`scipy.ndimage.binary_closing()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.binary_closing.html#scipy.ndimage.binary_closing)

Операция открытия удаляет небольшие структуры, в то время как операция закрытия заполняет небольшие отверстия. Таким образом, такие операции могут быть использованы для “очистки” изображения.

````python
a = np.zeros((50, 50))
a[10:-10, 10:-10] = 1
a += 0.25 * np.random.standard_normal(a.shape)
mask = a>=0.5
opened_mask = ndimage.binary_opening(mask)
closed_mask = ndimage.binary_closing(opened_mask)

imgs = [a, mask, opened_mask, closed_mask]
names = ['a', 'mask', 'opened_mask', 'closed_mask']
fig = plt.figure(figsize=(15,5))
for i in range(len(imgs)):
    fig.add_subplot(1,4,i+1)
    plt.imshow(imgs[i], cmap=plt.cm.gray)
    plt.title(names[i])
    plt.axis('off')
````

### **Упражнение**

Проверьте, что площадь восстановленного квадрата меньше площади исходного квадрата. (Обратное будет происходить, если шаг операции закрытия выполнить до шага операции открытия).

Для изображений в оттенках серого размывание (соответственно расширение) равносильно замене пикселя на минимальное (соответственно максимальное) значение среди пикселей, покрытых структурирующим элементом, центрированном на интересующем пикселе.

### Подключенные компоненты и измерения на изображениях

Давайте сначала сгенерируем хорошее синтетическое двоичное изображение.

````python
x, y = np.indices((100, 100))
sig = np.sin(2*np.pi*x/50.) * np.sin(2*np.pi*y/50.) * (1+x*y/50.**2)**2
mask = sig > 1
````

`scipy.ndimage.label()` присваивает разные метки каждому подключенному компоненту:

````python
labels, nb = ndimage.label(mask)
nb
````

````python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(sig)
plt.title('sig')
plt.axis('off')
plt.subplot(132)
plt.imshow(mask, cmap=plt.cm.gray)
plt.title('mask')
plt.axis('off')
plt.subplot(133)
plt.imshow(labels)
plt.title('labels')
plt.axis('off')
plt.show()
````

Теперь произведём измерения для каждого включённого элемента.  
Площади:

````python
areas = ndimage.sum(mask, labels, range(1, labels.max()+1))
areas
````

````python
maxima = ndimage.maximum(sig, labels, range(1, labels.max()+1))
maxima
````

Извлечём 4-й компонент и отбросим значения массив вокруг него:

````python
ndimage.find_objects(labels==4)
````

````python
sl = ndimage.find_objects(labels==4)
plt.imshow(sig[sl[0]]);
````
