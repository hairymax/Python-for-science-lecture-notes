## Специальные функции: [`scipy.special`](https://docs.scipy.org/doc/scipy/reference/special.html#module-scipy.special)

Специальные функции - функции встречающиеся в различных задачах математической физики, которые не выражаются через элементарные функции. Документации модуля `scipy.special` хорошо описана, поэтому мы не будем перечислять здесь все функции. Часто используемыми из них являются:

1. Функция Бесселя, такая как `scipy.special.jn()` (функция Бесселя n-го целого порядка);

3. Гамма-функция: `scipy.special.gamma()`, также обратите внимание на `scipy`.`special.gammaln()`, который даст логарифм гаммы с более высокой числовой точностью.
4. `Erf`, область под гауссовой кривой: `scipy.special.erf()`

## Линеная алгебра: [`scipy.linalg`](https://docs.scipy.org/doc/scipy/reference/linalg.html#module-scipy.linalg)

Модуль `scipy.linalg` предоставляет стандартные операции линейной алгебры, опираясь на базовую эффективную реализацию (известные пакеты Fortran: BLAS, LAPACK)

Функция [`scipy.linalg.det()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.det.html#scipy.linalg.det) вычисляет определитель квадратной матрицы:

````python
from scipy import linalg
arr = np.array([[1, 2],
                [3, 4]])
linalg.det(arr)
````

````python
arr = np.array([[3, 2],
                [6, 4]])
linalg.det(arr)
````

Для неквадратных матриц получим ошибку

````python
linalg.det(np.ones((3, 4)))
````

Функция [`scipy.linalg.inv()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.inv.html#scipy.linalg.inv) вычисляет матрицу обратную квадратной:

````python
arr = np.array([[1, 2],
                [3, 4]])
iarr = linalg.inv(arr)
iarr
````

````python
np.allclose(np.dot(arr, iarr), np.eye(2))
````

Наконец, вычисление обратной сингулярной матрицы (ее определитель равен нулю) вызовет ошибку `LinAlgError`:

````python
arr = np.array([[3, 2],
                [6, 4]])
linalg.inv(arr) 
````

Доступны более продвинутые операции, например разложение по сингулярным значениям (SVD):

````python
arr = np.arange(9).reshape((3, 3)) + np.diag([1, 0, 1])
uarr, spec, vharr = linalg.svd(arr)
````

Результирующий спектр массива имеет вид:

````python
spec 
````

Исходная матрица может быть перекомпилирована путем матричного умножения выходных данных `svd` через `np.dot`:

````python
sarr = np.diag(spec)
svd_mat = uarr.dot(sarr).dot(vharr)
np.allclose(svd_mat, arr)
````

SVD обычно используется в статистике и обработке сигналов. Многие другие стандартные разложения (QR, LU, Cholesky, Schur), а также решатели для линейных систем доступны в `scipy.linalg`.

## Интерполяция: [`scipy.interpolate`](https://docs.scipy.org/doc/scipy/reference/interpolate.html#module-scipy.interpolate)

`scipy.interpolate` полезен для интерполяции функции по экспериментальным данным, а также последующего получения значений в точках, для которых измерения отсутствуют. Модуль основан на подпрограммах Fortran - FITPACK 

Представим экспериментальные данные как набор точек, близких к синусоидальной функции, с добавлением случайного шума:

````python
measured_time = np.linspace(0, 1, 10)
noise = (np.random.random(10)*2 - 1) * 1e-1
measures = np.sin(2 * np.pi * measured_time) + noise
````

[`scipy.interpolate.interp1d`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html#scipy.interpolate.interp1d) может построить функцию линейной интерполяции:

````python
from scipy.interpolate import interp1d
linear_interp = interp1d(measured_time, measures)
````

Затем результат может быть вычислен в интересующий момент времени:

````python
interpolation_time = np.linspace(0, 1, 50)
linear_results = linear_interp(interpolation_time)
````

Кубическую интерполяцию также можно выбрать, указав необязательный аргумент ключевого слова kind:

````python
cubic_interp = interp1d(measured_time, measures, kind='cubic')
cubic_results = cubic_interp(interpolation_time)
````

Построим график

````python
plt.plot(measured_time, measures, 'o')
plt.plot(interpolation_time, linear_results)
plt.plot(interpolation_time, cubic_results)
plt.legend(('Измерения', 'Линейная интерполяция', 'Кубическая интерполяция'))
plt.show()
````

[`scipy.interpolate.interp2d`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp2d.html#scipy.interpolate.interp2d) аналогичен `scipy.interpolate.interp1d`, но для двумерных массивов. Обратите внимание, что для интерполяции точки должны оставаться в пределах диапазона заданных точек данных. Смотрите [Краткое упражнение по прогнозированию максимальной скорости ветра на станции Спроге](https://scipy-lectures.org/intro/summary-exercises/stats-interpolate.html#summary-exercise-stat-interp) для более подробного примера интерполяции сплайнов.

## Поиск экстремумов и приближение кривыми: [`scipy.optimize`](https://docs.scipy.org/doc/scipy/reference/optimize.html#module-scipy.optimize)

Под оптимизацией в англоязычной литературе понимаются проблемы численного нахождения экстремумов, поиска нулей.

Модуль [`scipy.optimize`](https://docs.scipy.org/doc/scipy/reference/optimize.html#module-scipy.optimize) предоставляет алгоритмы минимализации функций (скалярных или многомерных), приближения кривых и нахождения корней.

````python
from scipy import optimize
````

### Приближение кривыми

Допустим, у нас есть данные, представляющие синусоидальную волну с некоторым шумом:

````python
x_data = np.linspace(-5, 5, num=50)
y_data = 2.9 * np.sin(1.5 * x_data) + np.random.normal(size=50)

plt.plot(x_data, y_data, 'o');
````

Если мы знаем, что данные лежат на синусоидальной волне, но не знаем амплитуду или период, мы можем их найти методом наименьших квадратов. Сначала необходимо определить тестовую функцию, чтобы опробовать метод, например, здесь синусоида с неизвестной амплитудой `a` и периодом `b`:

````python
def test_func(x, a, b):
  return a * np.sin(b * x)
````

Воспользуемся [`spicy.optimize.curve_fit()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html#scipy.optimize.curve_fit), чтобы найти `a` и `b`:





````python
params, params_covariance = optimize.curve_fit(test_func, x_data, y_data, p0=[2, 2])
a, b = params
print(a, b)
````

````python
x_fit = np.linspace(-5, 5, num=100)
plt.plot(x_data, y_data, 'o')
plt.plot(x_fit, test_func(x_fit,a,b))
plt.legend(('Данные', 'Приближённая кривая'))
plt.show()
````

### **Упражнение: Приближение температурных данных кривыми**

Даны крайние (минимальные и максимальные) значения температур в Петропавловске-Камчатском для каждого месяца, начиная с января в градусах Цельсия:

### Нахождение минимума скалярной функции

Определим следующую функцию:

````python
def f(x):
  return x**2 + 10*np.sin(x)
````

И построим её график:

````python
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x))
plt.show()
````

Эта функция имеет глобальный минимум около значения $x = -1.3$ и локальный минимум около $3.8$.

Найти минимум можно с помощью `scipy.optimize.minimize()`, при вводе стартовой точки `x0` функция вернёт положение минимума, которое было найдено:



````python
result = optimize.minimize(f, x0=0)
result
````

````python
result.x # Координата минимума
````

**Методы:** Поскольку функция является гладкой, то тут продходят методы, основанные на градиентном понижении. [Алгоритм IBFGS](https://en.wikipedia.org/wiki/Limited-memory_BFGS) в принципе хорошо справляется с этой задачей:

````python
optimize.minimize(f, x0=0, method="L-BFGS-B")
````

Заметьте, что для нахождения хорошего значения минимума потребовалось всего 12 просчётов функций.

**Глобальный минимум:** Возможная проблема с этим подходом заключается в том, что если функция имеет локальные минимумы, алгоритм может найти их вместо глобального, в зависимости от начальной точки `x0`:

````python
res = optimize.minimize(f, x0=3, method="L-BFGS-B")
res.x
````

Если мы не знаем окрестность, в которой находится глобальный минимум, нам нужно прибегнуть к более затратной процедуре глобального поиска экстремума. Чтобы найти глобальный минимум, нам нужно использовать [`scipy.optimize.basinhopping()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.basinhopping.html#scipy.optimize.basinhopping) (добавлено в версии 0.12.0 Scipy). Эта функция комбинирует локальный поиск с выборкой стартовых точек:

````python
optimize.basinhopping(f, 0)
````

**Заметка**: `spipy` ранее содержал шаблон `anneal`, который был удалён в Scipy 0.16.0

**Ограничения:** Мы можем ограничить диапазон поиска минимума интервалом, например (0, 10), используя аргумент "bounds".

Поскольку `minimize()` работает в общем случае с многомерным массивом `x`, то аргумент "bounds" представляет собой список интервалов по каждому измерению.

````python
res = optimize.minimize(f, x0=1, bounds=((0, 10), ))
res.x
````

Подумайте, почему функция определяет значение 0, которое не является минимумом нашей функции.

## Статистика и случайные числа: [`scipy.stats`](https://docs.scipy.org/doc/scipy/reference/stats.html#module-scipy.stats)

Модуль `scipy.stats` содержит инструменты для работы со статистикой и вероятностного описания случайных процессов. Генератор случайных чисел для различных случайных процессов может быть найден в `numpy.random`.

### Распределения: гисторгамма и функция плотности вероятности

При наблюдении за событиями случайного процесса построенная гистограмма является оценкой функции протности вероятности (probability density function - PDF):

````python
samples = np.random.normal(size=5000)
bins = np.arange(-4, 4.1, 0.4)
bins
````

````python
histogram = np.histogram(samples, bins=bins, density=True)[0]
bins = 0.5*(bins[1:] + bins[:-1])
bins
````

````python
from scipy import stats
pdf = stats.norm.pdf(bins)
plt.bar(bins, histogram)
plt.plot(bins, pdf, color='red')
plt.legend(['Функция плотности\n вероятности', 'Гистограмма',], loc='upper right')
plt.show()
````

Если мы знаем, что случайный процесс принадлежит заданному семейству случайных процессов, таких как нормальные процессы, мы можем выполнить подбор наблюдений с максимальной вероятностью для оценки параметров основного распределения. Здесь мы подбираем процесс c нормальным распределением наблюдаемых данных:

````python
loc, std = stats.norm.fit(samples)
print(loc)
print(std)
````

### Статистические тесты

Статистический тест – это критерий принятия решения. Например, если у нас есть два набора наблюдений, которые с некоторым приблицением имеют  Гауссовское распределение, мы можем использовать t-критерий Стьюдента, чтобы определить, являются ли два набора наблюдений заметно различными. Например.

````python
plt.hist(np.random.normal(0, 1, size=1000), bins=25)
plt.hist(np.random.normal(1, 1, size=1000), bins=25)
plt.legend(['Набор 1', 'Набор 2'])
plt.show()
````

Для теста рассмотрим выборки поменьше

````python
a = np.random.normal(0, 1, size=100)
b = np.random.normal(1, 1, size=10)

stats.ttest_ind(a, b)
````

Итоговый вывод содержит:
* T-значение (`statistic`): число, знак которого пропорционален разнице между двумя разными процессами, а величина связана с значимостью этой разницы
* p-значение (`pvalue`): вероятность того, что два процесса являются одинаковыми. Если оно близко к 1, два процесса практически гарантированно являются одинаковыми. Чем ближе к нулю, тем больше вероятности, что процессы разные.
