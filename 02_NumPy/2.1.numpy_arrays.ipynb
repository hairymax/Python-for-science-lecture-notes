## Основные типы данных

Возможно, вы заметили, что в некоторых случаях элементы массива отображаются с точкой в конце (например, `2.` вместо `2`). Это связано с различием в используемом типе данных:

````python
a = np.array([1, 2, 3])
a.dtype
````

````python
b = np.array([1., 2., 3.])
b.dtype
````

Различные типы данных позволяют нам более компактно хранить данные в памяти, но чаще всего мы просто работаем с числами с плавающей запятой. Обратите внимание, что в приведенном выше примере NumPy автоматически определяет тип данных по введенным данным. 

Вы можете явно указать, какой тип данных вам нужен:

````python
c = np.array([1, 2, 3], dtype=float)
c.dtype
````

Тип данных **по умолчанию** - `float`:

````python
a = np.ones((3, 3))
a.dtype
````

Также NumPy работает и с другими типами:
- Complex
- Bool
- Strings
- и другие
    - int32
    - int64
    - uint32
    - uint64

````python
# Complex
d = np.array([1+2j, 3+4j, 5+6*1j])
print(d, d.dtype)
# Bool
e = np.array([True, False, False, True])
print(e, e.dtype)
# Strings
f = np.array(['Москва', 'Владивосток', 'Хабаровск'])
print(f, f.dtype)
````

## Индексация и срезы

К элементам массива можно обращаться и присваивать им значения так же, как и при работе с последовательностям Python (например, списками):

````python
a = np.arange(10)
a
````

````python
a[0], a[2], a[-1]
````

❗Индексы начинаются с 0, как и в других последовательностях Python (и C/C++).❗

Поддерживается обычная форма записи python для инвертирования последовательности:

````python
a[::-1]
````

Для многомерных массивов индексы представляют собой кортежи целых чисел:

````python
a = np.diag(np.arange(3))
a
````

````python
a[1, 1]
````

````python
a[2, 1] = 10 # third line, second column
a
````

````python
a[1]
````

""Примечание""
- В двумерном массиве первое измерение соответствует строкам, второе - столбцам.
- Для многомерного массива `a`, `a[0]` получается извлечением всех элементов в неуказанных измерениях.

**Срезы**. Из массивов, как и из других последовательностей Python, также могут быть получены срезы:

````python
a = np.arange(10)
a
````

````python
a[2:9:3] # [start:end:step]
````

Обратите внимание, что последний индекс не включен!

````python
a[:4]
````

Все три компонента числа в срезе указывать не обязательно: по умолчанию они равны: `start` равен 0, `end` - номеру последнего, а `step` = 1:

````python
a[1:3]
````

````python
a[::2]
````

````python
a[3:]
````

## Копии и представления

Операция извлечения среза создает представление исходного массиве, которое является просто способом доступа к данным массива. Таким образом, исходный массив не копируется в память.  
Можно использовать `np.may_share_memory()`, чтобы проверить, используют ли два массива один и тот же блок памяти. Однако обратите внимание, при выполнении этого метода могут быть ложные срабатывания.

**При изменении представления исходный массив также изменяется:**

````python
a = np.arange(10)
````

````python
b = a[::2]
b
````

````python
np.may_share_memory(a, b)
````

````python
b[0] = 12
b
````

````python
a
````

````python
a = np.arange(10)
c = a[::2].copy()  # принудительное копирование
c[0] = 12
a
````

````python
np.may_share_memory(a, c)
````

Такое поведение может удивлять по началу, но оно позволяет экономить и память, и время.

## Необычная индексация

В NumPy извлечь часть массива можно с помощью срезов, а также с помощью булевых или целочисленных массивов (масок). Этот метод называется *Fancy indexing* (англ. "Причудливая индексация"/"Необычная индексация"). Он создает копии, а не представления.

### Использование булевых масок

````python
np.random.seed(3)
a = np.random.randint(0, 21, 15)
a
````

````python
(a % 3 == 0)
````

````python
mask = (a % 3 == 0)
extract_from_a = a[mask] # можно и так:  a[a%3==0]
extract_from_a           # извлекаем подмассив маской
````

Индексайия по маске может быть очень полезной для присвоения нового значения подмассиву:

````python
a[a % 3 == 0] = -1
a
````

### Использование массива целых чисел

````python
a = np.arange(0, 100, 10)
a
````

Индексация может быть выполнено с массивом целых чисел, где один и тот же индекс повторяется несколько раз:

````python
a[[2, 3, 2, 4, 2]]  # заметка: [2, 3, 2, 4, 2] это список Python
````

При такой индексации можно присваивать новые значения:

````python
a[[9, 7]] = -100
a
````

Когда новый массив создается путем индексации с массивом целых чисел, новый массив имеет ту же форму, что и массив целых чисел:

````python
a = np.arange(10)
idx = np.array([[3, 4], [9, 7]])
print(idx.shape)
a[idx]
````
